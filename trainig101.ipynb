```python
## request devuelve un tipo byte se convierte cada linea a strgin para ponerla en una lista
## cuento los caracteres de cada linea y creo otra lista
s.iter_lines()
document =[]
ind = []
encoding = 'utf-8'
for c in s.iter_lines():
    document.append(c.decode(encoding))
    ind.append(len(c))

display(len(document))
display(len(ind))
```

```python
import pandas as pd
import requests

pd.set_option('display.max_columns', None)  

def readNews(url):
    s = requests.get(url)
    s.iter_lines()
    document =[]
    ind = []
    encoding = 'utf-8'
    for c in s.iter_lines():
        document.append(c.decode(encoding))
        ind.append(len(c))
    
    df = pd.DataFrame(list(zip(document, ind)), 
                   columns =['string', 'characters']) 
    df = df[df.characters >= 30]
    article =df[df['string'].str.contains('articleBody')]
    display(article.string)
```

```python
readNews("https://www.elcolombiano.com/colombia/vuelos-internacionales-en-colombia-seran-reactivados-el-19-de-septiembre-de-2020-LH13635608")
```
